```python
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
```

```python
model = Sequential(
    [
        # część konwolucyjna
        Conv2D(8, (3, 3), activation='relu', input_shape=(28, 28, 1)),
        MaxPooling2D((2, 2)),
        Dropout(0.25),

        # część "klasyczna"
        Flatten(),
        Dense(64, activation='relu'),
        Dense(10, activation='softmax')
    ]
)
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy', 'AUC'])
```

```python
X_train[0].shape
```

```python
y_train.shape
```

```python
history = model.fit(X_train, y_train, batch_size=512, epochs=30, validation_data = (X_test,y_test))
```

```python
def predict_cloth(url_link):
    urllib.request.urlretrieve(url_link, "X_pred")
    img = Image.open("X_pred")

    # Make the image square without distortion by padding or cropping
    size = max(img.width, img.height)
    padding = (size - img.width) // 2, (size - img.height) // 2
    img = ImageOps.expand(img, border=padding, fill='black')

    # Resize the square image to (28, 28)
    img = img.resize((28, 28))
    img.show()

    # Convert the image to grayscale if it's in RGB format
    if img.mode == 'RGB':
        img = img.convert('L')  # Convert to grayscale

    X_pred = np.array(img)
    X_pred = np.array([X_pred.reshape((28, 28, 1)) / 255.0])
    y_pred = model.predict(X_pred)
    predicted_classes = np.argmax(y_pred, axis=1)
    class_labels = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
    predicted_labels = [class_labels[class_index] for class_index in predicted_classes]
    return predicted_labels
```
